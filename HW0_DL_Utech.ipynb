```
import cv2
import numpy as np
```

```

from matplotlib import pyplot as plt

#Import image
image = cv2.imread('./owl.jpg')

#Show the image with matplotlib
plt.imshow(image)
plt.show()
```

```
cv2.rectangle(image, (1000,380), (1250, 600), (255,0,0), -1)
plt.imshow(image)
plt.show()
```

```
image = cv2.imread('./owl.jpg')

#kernel 3*3
kernel3 = np.ones((3, 3), np.float32) / 9
blurred = cv2.filter2D(image, -1, kernel3)

#kernel 5*5
kernel5 = np.ones((5,5), np.float32) /25
blurred1 = cv2.filter2D(image, -1, kernel5)

#kernel 7*7
kernel7 = np.ones((7, 7), np.float32) / 49
blurred2 = cv2.filter2D(image, -1, kernel7)

plt.imshow(blurred1)
plt.show()
```
